```python
#import libraries
import os
import pandas as pd
import re
import csv
from io import StringIO
from pathlib import Path
import numpy as np
```

```python
#read file
filename = Path(input("Enter the file name : "))
```

```python
#load csv to dataframe
col_list = ['desc', 'qty', 'price']
df = pd.read_csv(filename, usecols = [1,2,3], skiprows = 1, names = col_list)[:-4]
```

```python
#get rid of Pesos
df.price = df.price.str.replace('â‚±','')
```

```python
#extract content from description
df['content'] = df.desc.str.extract(r'(\d+k?g?)', expand = False).str.extract(r'(\d+)').fillna(1)
```

```python
#extract units from content
df['units'] = df.desc.str.extract(r'(\d+k?g?)', expand = False)
df['units'] = df.units.str.replace(r'(\d+)','').fillna('pcs')
df['units'] = df.units.replace(r'^\s*$',np.NaN, regex = True).fillna('pcs')
```

```python
#create item column containing basic item identifier
#remove Fresh
df['item'] = df.desc.str.replace('Fresh ','')
```

```python
df.item = df.item.str.replace(r'\".+\"','')
```

```python
#x = 'Fresh Bell Pepper "Red" (Siling Pari)250g'
pattern = r'(\(.+\))'
df.item = df.item.str.replace(pattern,'')
```

```python
pattern = r'(\d+k?g?)'
df.item = df.item.str.replace(pattern,'')
```

```python
#get rid of unneeded rows
df.date = pd.to_datetime(df.date, errors='coerce', dayfirst=True)
#df = df.loc[-df.date.isnull()]
```

```python
#convert to default date format
df.date.loc[:] = df.date.dt.strftime('%m/%d/%Y')
```

```python
#strip unneeded spaces
df.desc = df.desc.str.strip()
df.desc = df.desc.str.replace(" +"," ")
```

```python
#lower
df['desc'] = df.desc.str.lower()
```

```python
df.credit = df.credit.str.replace(' -','')
```

```python
#get rid of nans
df.debit = df.debit.fillna(0)
df.credit = df.credit.fillna(0)
```

```python
#export to csv
filename_csv_full = 'for import ' + filename.name
export_csv = df.to_csv (filename_csv_full, index = None, header=True)
```
